### 3. R√©cup√©rer le document HTML associ√© √† l‚Äôurl en utilisant `urllib.request.urlopen(url).read()`

```python
import urllib
page = urllib.request.urlopen(url).read()
page
```

### 4. Create a BeautifulSoup object, representing the structure of the HTML document. Use display the created object. You can use the prettify () function to get improved display.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page)
print(soup.prettify())
```

### 5. Display the title tag of the HTML document using `soup.title` and access the tag text using soup.title.string.

```python
print(soup.title)
print(soup.title.string)
```

### 6. Display the links contained in the HTML document (values ‚Äã‚Äãof href in the tags a).

```python
for anchor in soup.findAll('a', href=True):
    print(anchor['href'])
```
