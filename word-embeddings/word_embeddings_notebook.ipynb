### 4. Create a VizualizeEmbeddings function, which takes as input a model and displays the word vectors in two-dimensional space.

```python
# imports
from matplotlib import pyplot
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
def visualizeEmbeddings(model, caption):
  # get model vectors
  vects = model.wv.vectors
  # reduce vector dimension
  reduced_vect = pca.fit_transform(vects)
  # create scatterplot
  pyplot.scatter(reduced_vect[:,0], reduced_vect[:,1])
  pyplot.title(caption)
  words = list(model.wv.vocab)
  for i, word in enumerate(words):
    pyplot.annotate(word, xy=(reduced_vect[i,0], reduced_vect[i,1]))
  pyplot.show()
```

### 5. Use the `VizualizeEmbeddings` function to view the vectors generated by the two models: modelwv and modelft. Compare

```python
title1 = "Visualizing Word2Vec embeddings"
visualizeEmbeddings(modelwv, title1)
```

```python
title2 = "Visualizing FastText embeddings"
visualizeEmbeddings(modelft, title2)
```

```python
import glove
# from glove import Corpus, Glove

# create glove corpus
gloveCorpus = glove.Corpus()

# fit on sentences
gloveCorpus.fit(sentences)

# word embedding
word_embedding = glove.Glove(no_components=50, learning_rate=0.05)

# fit word embedding
word_embedding.fit(gloveCorpus.matrix)

# supply dictionary
word_embedding.add_dictionary(gloveCorpus.dictionary)
```
